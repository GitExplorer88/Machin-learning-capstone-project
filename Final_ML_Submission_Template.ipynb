### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0: There is no impact of Annual_premium on Customer's Response.

H1: Annual_premium amount have impact on customer's Response.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
alpha=0.5
from scipy.stats import mannwhitneyu
premium_interest=df[df['Response']==1]['Annual_Premium']
premium_no_interest=df[df['Response']==0]['Annual_Premium']
statistic,p_value=mannwhitneyu(premium_interest,premium_no_interest)
if p_value>alpha:
  print('Null Hypothesis is proved')
elif p_value<=alpha:
  print('Alternate Hypothesis have strong evidences')
````

### 2. Handling Outliers

````python
#creating functions for outlier detection
def skew_dist_out_detection(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]
    if len(outliers) == 0:
        print(column, 'column has no outliers')
    else:
        print(column, 'column has outliers')

def norm_dist_out_detection(df, column, threshold=3):
    from scipy.stats import zscore
    z = zscore(df[column])
    outliers = df[abs(z) > threshold]
    if len(outliers) == 0:
        print(column, 'column has no outliers')
    else:
        print(column, 'column has outliers')
````

````python
#those columns who follow skewed distribution ,checking that their have any outliers
skew_dist_out_detection(df,'Age')
skew_dist_out_detection(df,'Annual_Premium')
skew_dist_out_detection(df,'Policy_Sales_Channel')
#those columns who follow normal distribution,checking that their have any outliers
norm_dist_out_detection(df,'Region_Code')
norm_dist_out_detection(df,'Vintage')
````
