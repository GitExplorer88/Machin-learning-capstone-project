## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#checking that our dataset have imbalanced nature or not
response_types_count=df['Response'].value_counts().reset_index()
response_types_count
````

````python
#checking that how much customers comes for insurance on the basis of their vehicle damage(in the past) information
response_based_on_vehicledamage=df.groupby(['Vehicle_Damage','Response']).size().reset_index(name='count')
response_based_on_vehicledamage
````

````python
#checking impact of annual premium amount on customers interest
impact_of_premiumammount=df.groupby('Response')['Annual_Premium'].mean().reset_index()
impact_of_premiumammount
````

````python
#checking vehicle age's impact
impact_of_vehicle_age=df.groupby(['Vehicle_Age','Vehicle_Damage']).size().reset_index(name='count')
impact_of_vehicle_age
````

````python
#vehicles age impact on response
vehicle_age_impact_response=df.groupby(['Vehicle_Age','Response']).size().reset_index(name='count')
vehicle_age_impact_response
````

````python
#checking which gender have most contribution in our customers
gender_contribution=df['Gender'].value_counts().reset_index(name='counts')
gender_contribution
````

````python
#checking which age group has more interest in insurance policy
age_group=df.groupby('Response')['Age'].mean().reset_index()
age_group
````

````python
#checking that how much previouse customers interest in insurance policy
previouse_customers_interest=df.groupby('Previously_Insured')['Response'].sum().reset_index()
previouse_customers_interest
````

````python
#checking that drivince lincense can create any impact on customers interest
driving_license_impact=df.groupby('Driving_License')['Response'].sum().reset_index()
driving_license_impact
````

````python
#checking that vintage can do impact on customers choice
vintage_impact=df.groupby('Response')['Vintage'].mean().reset_index()
vintage_impact
````
